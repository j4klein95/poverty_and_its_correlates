```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df1 = pd.read_csv("../csv/city_marriage_2016.csv", encoding="latin-1")
df2 = pd.read_csv("../csv/city_pov.csv", encoding="latin-1")
df3 = pd.read_csv("../csv/city_pov_2016.csv", encoding="ISO-8859-1")
df4 = pd.read_csv("../csv/city_pov_2016_perc.csv", encoding="latin-1")
df5 = pd.read_csv("../csv/city_pov_v_health.csv", encoding="latin-1")
df6 = pd.read_csv("../csv/city_pov_2016_mn.csv")

df3 = df3.replace("N", np.nan)
df3 = df3.fillna('0')
```

```python
plt.subplots(figsize=(20, 15))
ax = plt.axes()
corr = df1.corr()

sns.heatmap(corr,
            annot=True,
           xticklabels=corr.columns.values,
           yticklabels=corr.columns.values,
          cmap = plt.cm.plasma)
```
